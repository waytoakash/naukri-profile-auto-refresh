```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.chrome.service import Service
import time
import os
from dotenv import load_dotenv

# Change directory to where your .env file is located
os.chdir(r"D:\Resume\naukri_auto_refresh_V2")

# Now load the .env file
load_dotenv()

# Fetch credentials
EMAIL = os.getenv("NAUKRI_EMAIL")
PASSWORD = os.getenv("NAUKRI_PASSWORD")
CHROME_DRIVER_PATH = os.getenv("CHROME_DRIVER_PATH")

# Show results
print("ðŸ“§ Email loaded:", EMAIL)
print("ðŸ“§ Password loaded:", PASSWORD)
print("ðŸ“‚ ChromeDriver path:", CHROME_DRIVER_PATH)
```

```python
# Set up Chrome driver
service = Service(CHROME_DRIVER_PATH)
driver = webdriver.Chrome(service=service)
```

```python
# Open Naukri
driver.get("https://www.naukri.com/")
driver.maximize_window()
time.sleep(3)

# Click Login
login_button = driver.find_element(By.LINK_TEXT, "Login")
login_button.click()
time.sleep(3)

from selenium.webdriver.common.by import By
import time

# Enter credentials
driver.find_element(By.XPATH, "//input[@type='text' or @name='email']").send_keys(EMAIL)
driver.find_element(By.XPATH, "//input[@type='password']").send_keys(PASSWORD)

driver.find_element(By.XPATH, "//button[contains(text(), 'Login')]").click()

time.sleep(5)

# ðŸš€ Redirect to profile page explicitly
driver.get("https://www.naukri.com/mnjuser/profile")

# Wait for profile page to load fully
time.sleep(5)

import random
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.webdriver.common.by import By
from datetime import datetime

wait = WebDriverWait(driver, 15)

headlines = [
    "Data & Analytics Professional | Skilled in Advanced Analytics, Automation & Business Intelligence | Data-Driven Decision Making & Process Optimization | Stakeholder Management | Team Handling | Immediate Joiner | Watch: https://bit.ly/MyIntro_Video",
    "I am a immediate joiner & looking for Business Analyst/ Sr. Data Analyst role. Experienced Professional with 7 Years in Sql, Tableau, Looker Studio, Python, Excel, Google Sheet, Tableau, Google Big Query, Data Lakes, SAP HANA, Stakeholder management, Team handling."
]

try:
    upload_input = wait.until(
        EC.presence_of_element_located((By.XPATH, "//input[@type='file']"))
    )
    upload_input.send_keys(r"D:\Resume\My_CV_202511.pdf")  # <- Replace with your actual resume path 
    print("ðŸ“Ž Resume re-uploaded.")

    # â³ Wait a bit to simulate activity
    time.sleep(5)

    print(f"âœ… Profile refreshed successfully at {datetime.now().strftime('%Y-%m-%d %H:%M:%S')}")

except Exception as e:
    print("âš ï¸ Could not refresh profile:", e)

# === Finish ===
time.sleep(5)
driver.quit()
print("âœ… Script finished")
```
